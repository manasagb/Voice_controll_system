```python
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
import speech_recognition as sr
from tensorflow.keras.preprocessing import image
```

```python
while(True):
    r=sr.Recognizer()
    try:
        with sr.Microphone() as source:
            print("say..")
            audio=r.listen(source)
            print("done..")
        input=r.recognize_google(audio)
        if ('not' in input or "don't" in input or 'dont' in input) and ('chrome' in input or 'notepad' in input or 'vlc' in input or 'km' in input or 'player' in input or 'editor' in input):
            print("Ok fine :)")
        elif ('notepad' in input or 'editor' in input) and ('run' in input or 'execute' in input or 'open' in input):
            os.system('notepad')
        elif('chrome' in input or 'browser' in input) and ('run' in input or 'execute' in input or 'open' in input):
            os.system('chrome')
        elif('VLC' in input or 'VLC player' in input) and ('run' in input or 'execute' in input or 'open' in input):
            os.system('start vlc')
        elif('KM' in input or 'KM player' in input) and ('run' in input or 'execute' in input or 'open' in input):
            os.system('start kmplayer')
        elif('edge' in input or 'microsoft' in input or 'msedge' in input) and ('run' in input or 'execute' in input or 'open' in input):
            os.system('start msedge')
        elif('mozilla' in input or 'firefox' in input) and ('run' in input or 'execute' in input or 'open' in input):
            os.system('start firefox')
        elif('image' in input or 'photo' in input) and ('capture' in input or 'click' in input):
            vco=cv2.VideoCapture(0)
            res=True
            while res:
                ret,frame=vco.read()
                cv2.imwrite('new.jpg',frame)
                res=False
            img1=image.load_img('new.jpg')
            vco.release()
            cv2.destroyAllWindows()
        elif('show' in input or 'display' in input) and ('my' in input) and ('image' in input or 'photo' in input):
            try:
                plt.imshow(img1)
            except:
                print("Still you have not captured image")
        elif('exit' in input or 'quit' in input):
            print("Thank You:)")
            break
        else:
            print("I'm not getting")
    except:
        print('say something..')
```
